```python
import transformers
from transformers import pipeline

input_text="Natural Language Processing (NLP) enables machine learning algorithms to organize and understand human language. NLP enables machines to not only gather text and speech but also identify the core meaning it should respond to. Human language is complex, and constantly evolving, which means natural language processing has quite the challenge. Tokenization is one of the many pieces of the puzzle in how NLP works."
print(input_text)
```

```python
summarizer=pipeline("summarization")
summarizer(f"Text input: {input_text}",max_length=30,min_length=10,do_sample=False)
```

```python
nlp=pipeline("ner")
example="Obama is going to New York"

ner_results=nlp(example)
print(ner_results)
```
